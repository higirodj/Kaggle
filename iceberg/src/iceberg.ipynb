```python
def load_iceberg_data(dataset_path=DATA_PATH):
    """
    Read the training and testing datasets from JSON into a dataframe. 
    
    Arguments: 
    dataset_path - path to the datasets.
    
    Returns: 
    2 dataframes containing training and testing datasets.
    """
    json_training_path = os.path.join(dataset_path, "train.json")
    json_test_path = os.path.join(dataset_path, "test.json")
    
    return pd.read_json(json_training_path), pd.read_json(json_test_path)
```

```python
def get_images(data):
    """ 
    Reshape each image into a 75x75x3 (with 3-channels) for use in neural networks.
    
    Arguments:
    data - dataframe containing images.
    
    Returns:
    an array of images with the following dimensions (75x75x3).
    """
    band1 = np.array([np.array(item).astype(np.float32).reshape(75, 75)
                      for item in data["band_1"]])
    band2 = np.array([np.array(item).astype(np.float32).reshape(75, 75)
                      for item in data["band_2"]])
    band3 = np.mean(np.array([band1, band2]), axis = 0)
    
    band1 = (band1 + abs(band1.min())) / np.max((band1 + abs(band1.min())))
    band2 = (band2 + abs(band2.min())) / np.max((band2 + abs(band2.min())))
    band3 = (band3 + abs(band3.min())) / np.max((band3 + abs(band3.min())))
    
    return np.concatenate([band1[:, :, :, np.newaxis],
                           band2[:, :, :, np.newaxis],
                           band3[:, :, :, np.newaxis]], axis=-1)    
```

```python
def process_images(X, Y, whitening):
    """
    Generate batches of images for augmentation.
    
    Arguments: 
    X - input image. 
    Y - input label.
    whitening - boolean variable for selecting zca_whitening.
    
    Returns: an image data generator.
    """
    datagen = ImageDataGenerator(
        zca_whitening=whitening,
        zca_epsilon=1e-6,
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')
    datagen.fit(X_train)   
    generator = datagen.flow(X, Y, batch_size = batch_size, shuffle = False, seed = seed)
    
    return generator
```

```python
def get_callbacks():
    """
    Provides view on internal states and statistics of the model during training.
    
    Arguments: 
    none.
    
    Returns: 
    useful metrics.
    """
    model_checkpoint = ModelCheckpoint(OUTPUT_PATH+'weights.{epoch:02d}-{val_loss:.2f}.hdf5',
                                   monitor = 'val_loss', verbose = 1, save_best_only = True,
                                   save_weights_only = False, mode = 'auto', period = 1)
    early_stopping = EarlyStopping(monitor='val_loss', min_delta=0,
                               patience=10, verbose=1, mode='auto')
    callbacks = [model_checkpoint, early_stopping]
    
    return callbacks
```

## 1a. VGG16 Functions

```python
def VGG16_MODEL(X_train, Y_train, X_test, Y_test, augmentation = False, whitening=False):
    """
    Generates pretrained VGG16 model that will be used for the classifier.
    
    Arguments:
    X_train - input training images.
    Y_train - input training labels.
    X_test - input validation. 
    Y_test - input validation labels.
    augmentation - boolean variable for selecting image augmentation.
    whitening - boolean variable for selecting zca_whitening.
    
    Returns: useful metrics.
    """
    base = VGG16(include_top = False, weights = 'imagenet', input_shape = X_train.shape[1:])
    x = GlobalAveragePooling2D(name='global_avg_pool')(base.output)
    x = Dense(4096, activation='relu', name='fc1')(x)
    x = Dropout(0.5)(x) 
    x = Dense(4096, activation='relu', name='fc2')(x)
    x = Dense(1, activation='sigmoid', name='predictions')(x)
    
    model = Model(inputs = base.input, outputs = x)
    model.summary()
    
    for layer in base.layers:
        layer.trainable = False
    model.compile(loss = 'binary_crossentropy',
              optimizer=SGD(lr=1e-4, momentum=0.9),
              metrics=['accuracy'])
    model.summary()
    
    if not augmentation:
        print('[IMAGE AUGMENTATION NOT SELECTED]')
        model.fit(X_train, Y_train,
                  epochs = epochs,
                  batch_size = batch_size,          
                  validation_data=(X_test, Y_test),
                  callbacks = get_callbacks())
    else:
        print('[IMAGE AUGMENTATION SELECTED]')
        generator = process_images(X_train, Y_train, whitening)
        model.fit_generator(generator, 
                            epochs = epochs,
                            steps_per_epoch = X_train.shape[0] // batch_size,
                            validation_data=(X_test, Y_test),
                            callbacks = get_callbacks())
        # Weights to be used in fine-tuning a classifier
        model.save_weights(OUTPUT_PATH + 'learned_weights.h5') 
        
    return model
```

## 1b. ResNet50 Functions

```python
def identity_block(X, f, filters, stage, block):
    """
    Arguments:
    X -- input tensor of shape (m, n_H_prev, n_W_prev, n_C_prev)
    f -- integer, specifying the shape of the middle CONV's window for the main path
    filters -- python list of integers, defining the number of filters in the CONV layers of the main path
    stage -- integer, used to name the layers, depending on their position in the network
    block -- string/character, used to name the layers, depending on their position in the network
    
    Returns:
    X -- output of the identity block, tensor of shape (n_H, n_W, n_C)
    """
    
    # defining name basis
    conv_name_base = 'res' + str(stage) + block + '_branch'
    bn_name_base = 'bn' + str(stage) + block + '_branch'
    
    # Retrieve Filters
    F1, F2, F3 = filters
    
    # Save the input value. You'll need this later to add back to the main path. 
    X_shortcut = X
    
    # First component of main path
    X = Conv2D(filters = F1, kernel_size = (1, 1), strides = (1,1), padding = 'valid', name = conv_name_base + '2a', 
               kernel_initializer = glorot_uniform(seed=0))(X)
    X = BatchNormalization(axis = 3, name = bn_name_base + '2a')(X)
    X = Activation('relu')(X)
    
    # Second component of main path
    X = Conv2D(filters = F2, kernel_size = (f, f), strides = (1,1), padding = 'same', name = conv_name_base + '2b', 
               kernel_initializer = glorot_uniform(seed=0))(X)
    X = BatchNormalization(axis = 3, name = bn_name_base + '2b')(X)
    X = Activation('relu')(X)

    # Third component of main path
    X = Conv2D(filters = F3, kernel_size = (1, 1), strides = (1,1), padding = 'valid', name = conv_name_base + '2c', 
               kernel_initializer = glorot_uniform(seed=0))(X)
    X = BatchNormalization(axis = 3, name = bn_name_base + '2c')(X)

    # Final step: Add shortcut value to main path, and pass it through a RELU activation
    X = Add()([X_shortcut, X])
    X = Activation('relu')(X)

    return X
```

```python
def convolutional_block(X, f, filters, stage, block, s = 2):
    """
    Arguments:
    X -- input tensor of shape (m, n_H_prev, n_W_prev, n_C_prev)
    f -- integer, specifying the shape of the middle CONV's window for the main path
    filters -- python list of integers, defining the number of filters in the CONV layers of the main path
    stage -- integer, used to name the layers, depending on their position in the network
    block -- string/character, used to name the layers, depending on their position in the network
    s -- Integer, specifying the stride to be used
    
    Returns:
    X -- output of the convolutional block, tensor of shape (n_H, n_W, n_C)
    """
    
    # defining name basis
    conv_name_base = 'res' + str(stage) + block + '_branch'
    bn_name_base = 'bn' + str(stage) + block + '_branch'
    
    # Retrieve Filters
    F1, F2, F3 = filters
    
    # Save the input value
    X_shortcut = X


    ##### MAIN PATH #####
    # First component of main path 
    X = Conv2D(F1, (1, 1), strides = (s,s), name = conv_name_base + '2a', kernel_initializer = glorot_uniform(seed=0))(X)
    X = BatchNormalization(axis = 3, name = bn_name_base + '2a')(X)
    X = Activation('relu')(X)

    # Second component of main path
    X = Conv2D(filters = F2, kernel_size = (f, f), strides = (1,1), padding = 'same', name = conv_name_base + '2b', 
               kernel_initializer = glorot_uniform(seed=0))(X)
    X = BatchNormalization(axis = 3, name = bn_name_base + '2b')(X)
    X = Activation('relu')(X)

    # Third component of main path
    X = Conv2D(filters = F3, kernel_size = (1, 1), strides = (1,1), padding = 'valid', name = conv_name_base + '2c', 
               kernel_initializer = glorot_uniform(seed=0))(X)
    X = BatchNormalization(axis = 3, name = bn_name_base + '2c')(X)

    ##### SHORTCUT PATH ####
    X_shortcut = Conv2D(filters = F3, kernel_size = (1, 1), strides = (s,s), padding = 'valid', name = conv_name_base + '1', 
               kernel_initializer = glorot_uniform(seed=0))(X_shortcut)
    X_shortcut = BatchNormalization(axis = 3, name = bn_name_base + '1')(X_shortcut)

    # Final step: Add shortcut value to main path, and pass it through a RELU activation
    X = Add()([X_shortcut, X])
    X = Activation('relu')(X)
    
    return X
```

```python
def ResNet50(input_shape = (75, 75, 3), classes = 2):
    """
    Implementation of the popular ResNet50 the following architecture:
    CONV2D -> BATCHNORM -> RELU -> MAXPOOL -> CONVBLOCK -> IDBLOCK*2 -> CONVBLOCK -> IDBLOCK*3
    -> CONVBLOCK -> IDBLOCK*5 -> CONVBLOCK -> IDBLOCK*2 -> AVGPOOL -> TOPLAYER

    Arguments:
    input_shape -- shape of the images of the dataset
    classes -- integer, number of classes

    Returns:
    model -- a Model() instance in Keras
    """
    
    # Define the input as a tensor with shape input_shape
    X_input = Input(input_shape)

    
    # Zero-Padding
    X = ZeroPadding2D((3, 3))(X_input)
    
    # Stage 1
    X = Conv2D(64, (7, 7), strides = (2, 2), name = 'conv1', kernel_initializer = glorot_uniform(seed=0))(X)
    X = BatchNormalization(axis = 3, name = 'bn_conv1')(X)
    X = Activation('relu')(X)
    X = MaxPooling2D((3, 3), strides=(2, 2))(X)

    # Stage 2
    X = convolutional_block(X, f = 3, filters = [75, 75, 300], stage = 2, block='a', s = 1)
    X = identity_block(X, 3, [75, 75, 300], stage=2, block='b')
    X = identity_block(X, 3, [75, 75, 300], stage=2, block='c')
    
    # Stage 3 
    X = convolutional_block(X, f = 3, filters = [150, 150, 600], stage = 3, block='a', s = 2)
    X = identity_block(X, 3, [150, 150, 600], stage=3, block='b')
    X = identity_block(X, 3, [150, 150, 600], stage=3, block='c')
    X = identity_block(X, 3, [150, 150, 600], stage=3, block='d')

    # Stage 4
    X = convolutional_block(X, f = 3, filters = [300, 300, 1200], stage = 4, block='a', s = 2)
    X = identity_block(X, 3, [300, 300, 1200], stage=4, block='b')
    X = identity_block(X, 3, [300, 300, 1200], stage=4, block='c')
    X = identity_block(X, 3, [300, 300, 1200], stage=4, block='d')
    X = identity_block(X, 3, [300, 300, 1200], stage=4, block='e')
    X = identity_block(X, 3, [300, 300, 1200], stage=4, block='f')

    # Stage 5
    X = convolutional_block(X, f = 3, filters = [600, 600, 2400], stage = 5, block='a', s = 2)
    X = identity_block(X, 3, [600, 600, 2400], stage=5, block='b')
    X = identity_block(X, 3, [600, 600, 2400], stage=5, block='c')

    # AVGPOOL
    X = AveragePooling2D(pool_size=(2,2), name='avg_pool')(X)

    # output layer
    X = Flatten()(X)
    X = Dense(classes, activation='softmax', name='fc' + str(classes), kernel_initializer = glorot_uniform(seed=0))(X)
    
    # Create model
    model = Model(inputs = X_input, outputs = X, name='ResNet50')

    return model
```
